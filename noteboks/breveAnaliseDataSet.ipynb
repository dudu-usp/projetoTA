##Daily dataset do block 0

```python
import pandas as pd

real_path ="/root/.cache/kagglehub/datasets/jeanmidev/smart-meters-in-london/versions/21/daily_dataset/daily_dataset/block_0.csv"

df = pd.read_csv(real_path)
```

Aqui temos:

LCLid: id de cada contador de energia

day: dia da medicao

demais: valores relacionados a cada dia da medicao

```python
df['LCLid'].value_counts()
```

```python
len(df['LCLid'].unique())
```

```python
df['LCLid'].value_counts().mean()
```

A analise de 1 quarteirao (block) nois indica que temos 50 contadores (cada contador eh identificado porum LCLid). E cada contador tem, em media, dados de 500 dias

```python
df['day'] = pd.to_datetime(df['day'])


range_datas = df.groupby('LCLid')['day'].agg(['min', 'max'])

range_datas
```

```python
range_datas.describe()
```

```python
range_datas.value_counts()
```

Como vemos, o range de datas nao eh uniforme, ou seja, temos diferentes medidores que fizeram medicoes em periodos distintos
